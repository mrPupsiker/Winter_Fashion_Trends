A. Загрузка и первичный обзор

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Winter_Fashion_Trends_Dataset.csv")

df.head(50)
```

```python
df.describe()
```

```python
df.info
```

```python
df['Season_Type'] = df['Season'].apply(lambda x: x.split(' ')[0]).astype('category')
df['Season_Year'] = df['Season'].apply(lambda x: int(x.split(' ')[1]))
df = df.drop('Season', axis=1)

for col in ['Brand', 'Category', 'Color', 'Material', 'Style', 'Gender', 'Trend_Status']:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = df[col].astype('category')

print("\nОбновленные типы данных:")
df.info()
```

B. Обработка данных

```python
duplicates = df.duplicated()
print(f"Всего дубликатов: {duplicates.sum()}")
duplicated_rows = df[duplicates]
print(duplicated_rows)
```

```python
from scipy import stats
threshold = 3
for column in df.select_dtypes(include=[np.number]).columns:
  z_scores = np.abs(stats.zscore(df[column], nan_policy='omit'))
  outliers = np.where(z_scores > threshold)[0]
  print(outliers)
```

```python
print(df.describe)
```

```python
for col in df.columns:
  print(f"\n--- Столбец: {col} ---")
  print(f"Тип данных: {df[col].dtype}")
  print(f"Всего значений: {len(df[col])}")
  print(f"Уникальных значений: {df[col].nunique()}")
  print(f"Пропусков: {df[col].isnull().sum()} ({df[col].isnull().sum()/len(df)*100:.1f}%)")
  if df[col].nunique() <= 20:
    print("\nРаспределение значений:")
    freq = df[col].value_counts(dropna=False)
    print(freq)
    # Процентное распределение
    print("\nПроцентное распределение:")
    perc = df[col].value_counts(dropna=False, normalize=True) * 100
    print(perc.round(2))
    # Если много уникальных значений, показываем топ-10
  else:
    print(f"\nТоп-10 самых частых значений:")
    top_10 = df[col].value_counts().head(10)
    print(top_10)
    print(f"\nТоп-10 самых редких значений:")
    bottom_10 = df[col].value_counts().tail(10)
    print(bottom_10)
```

```python
method='pearson'
numeric_df = df.select_dtypes(include=[np.number])
# Вычисляем корреляционную матрицу
corr_matrix = numeric_df.corr(method=method)
print("\nКорреляционная матрица:")
print(corr_matrix.round(3))
# Находим самые сильные корреляции
print(f"\nСАМЫЕ СИЛЬНЫЕ КОРРЕЛЯЦИИ:")
# Создаем список всех пар корреляций
corr_pairs = []
for i in range(len(corr_matrix.columns)):
  for j in range(i+1, len(corr_matrix.columns)):
    col1 = corr_matrix.columns[i]
    col2 = corr_matrix.columns[j]
    corr_value = corr_matrix.iloc[i, j]
    if not np.isnan(corr_value):
      corr_pairs.append({
        'var1': col1,
        'var2': col2,
        'correlation': corr_value,
        'abs_correlation': abs(corr_value)})
# Сортируем по абсолютному значению корреляции
    corr_pairs_df = pd.DataFrame(corr_pairs)
    if not corr_pairs_df.empty:
        corr_pairs_df = corr_pairs_df.sort_values('abs_correlation', ascending=False)
        print("\nТоп-10 самых сильных корреляций:")
        print(corr_pairs_df.head(10).to_string(index=False))
        print(f"\nКоличество сильных корреляций (|r| > 0.7): {len(corr_pairs_df[corr_pairs_df['abs_correlation'] > 0.7])}")
        print(f"Количество умеренных корреляций (0.5 < |r| < 0.7): {len(corr_pairs_df[(corr_pairs_df['abs_correlation'] > 0.5) & (corr_pairs_df['abs_correlation'] <= 0.7)])}")
        print(f"Количество слабых корреляций (|r| < 0.5): {len(corr_pairs_df[corr_pairs_df['abs_correlation'] <= 0.5])}")
```

C. Визуализация данных

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Устанавливаем стиль
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Для русских подписей
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# График 1: Гистограмма распределения продаж
plt.figure(figsize=(12, 8))

# Создаем subplot с 2x2 сеткой
plt.subplot(2, 2, 1)
plt.hist(df['Price(USD)'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Распределение цен', fontsize=14, fontweight='bold')
plt.xlabel('Цены', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# График 2: Линейный график популярности в зависимости от цен
plt.figure(figsize=(14, 6))

# Отсортируем данные по цене для осмысленного линейного графика
df_sorted_by_price = df.sort_values(by='Price(USD)')

# Создаем subplot
plt.subplot(1, 1, 1) # Изменил на 1,1,1 так как будет один график на всю фигуру

# Основной график: популярность по цене
plt.plot(df_sorted_by_price['Price(USD)'], df_sorted_by_price['Popularity_Score'],
         marker='o', markersize=4, linewidth=1, color='blue', alpha=0.5, label='Популярность')

# Добавляем скользящее среднее для Popularity_Score
window = 10 # Увеличил окно для более гладкой линии, можно настроить
rolling_mean_popularity = df_sorted_by_price['Popularity_Score'].rolling(window=window).mean()
plt.plot(df_sorted_by_price['Price(USD)'], rolling_mean_popularity,
         linewidth=3, color='red', label=f'Скользящее среднее популярности ({window} точек)')

plt.title('Динамика популярности в зависимости от цен', fontsize=16, fontweight='bold')
plt.xlabel('Цена (USD)', fontsize=12)
plt.ylabel('Оценка популярности', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout() # Улучшаем расположение элементов
plt.show()
```

```python
# График: Столбчатая диаграмма 'Популярность по брендам'
plt.figure(figsize=(14, 8))

# Группируем по 'Brand' и вычисляем среднюю 'Popularity_Score'
popularity_by_brand = df.groupby('Brand')['Popularity_Score'].mean().sort_values(ascending=False)

bars = plt.bar(popularity_by_brand.index, popularity_by_brand.values,
               color=plt.cm.viridis(np.linspace(0, 1, len(popularity_by_brand))), # Используем цветовую палитру
               edgecolor='black')

plt.title('Средняя популярность по брендам', fontsize=16, fontweight='bold')
plt.xlabel('Бренд', fontsize=12)
plt.ylabel('Средняя оценка популярности', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45, ha='right') # Поворачиваем метки для лучшей читаемости

# Добавляем значения на столбцы
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.05, # Небольшой отступ сверху
             f'{height:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout() # Улучшаем расположение элементов
plt.show()
```

```python
# График 4: Boxplot (ящики с усами) для цен
plt.figure(figsize=(14, 10))

# Подграфик 1: Boxplot для Price(USD)
plt.subplot(2, 2, 1)
box = plt.boxplot(df['Price(USD)'], vert=True, patch_artist=True,
                  boxprops=dict(facecolor='lightblue', color='black'),
                  medianprops=dict(color='red', linewidth=2),
                  whiskerprops=dict(color='black'),
                  capprops=dict(color='black'),
                  flierprops=dict(marker='o', markerfacecolor='red',
                                  markersize=5, alpha=0.5))

plt.title('Boxplot цен (USD)', fontsize=14, fontweight='bold')
plt.ylabel('Цена (USD)', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Добавляем статистику
stats = df['Price(USD)'].describe()
plt.text(0.7, 0.95,
         f"Q1: {stats['25%']:.0f}\nQ3: {stats['75%']:.0f}\nIQR: {stats['75%']-stats['25%']:.0f}\nВыбросов: {len(df[(df['Price(USD)'] < stats['25%'] - 1.5*(stats['75%']-stats['25%'])) | (df['Price(USD)'] > stats['75%'] + 1.5*(stats['75%']-stats['25%']))])}",
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
```

```python
# График 5: Heatmap корреляций
plt.figure(figsize=(14, 10))

# Подграфик 1: Базовая heatmap
plt.subplot(2, 2, 1)

# Вычисляем корреляционную матрицу
corr_matrix = df.select_dtypes(include=[np.number]).corr()

# Создаем heatmap
im = plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto')

plt.title('Корреляционная матрица (числовые переменные)', fontsize=14, fontweight='bold')
plt.colorbar(im, fraction=0.046, pad=0.04)

# Устанавливаем метки
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

# Добавляем значения в ячейки
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        text = plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}',
                       ha="center", va="center",
                       color="white" if abs(corr_matrix.iloc[i, j]) > 0.5 else "black",
                       fontsize=10)
```

D. Группировки и агрегации

```python
avg_price_by_brand = df.groupby('Brand')['Price(USD)'].mean().sort_values(ascending=False)
print("\nСредняя цена по брендам:")
print(avg_price_by_brand)
```

```python
avg_price_by_brand = df.groupby('Brand', observed=False)['Price(USD)'].mean().sort_values(ascending=False)
print("\nСредняя цена по брендам:")
print(avg_price_by_brand)
```

```python
item_count_by_category = df.groupby('Category', observed=False).size().sort_values(ascending=False)
print("\nКоличество товаров по категориям:")
print(item_count_by_category)
```

```python
popularity_by_gender = df.groupby('Gender', observed=False)['Popularity_Score'].sum().sort_values(ascending=False)
print("\nСуммарная оценка популярности по полу:")
print(popularity_by_gender)
```

```python
avg_price_pivot = df.pivot_table(values='Price(USD)', index='Brand', columns='Gender', aggfunc=np.mean)
print("\nСводная таблица средней цены по бренду и полу:")
print(avg_price_pivot)
```

```python
avg_price_pivot = df.pivot_table(values='Price(USD)', index='Brand', columns='Gender', aggfunc='mean', observed=False)
print("\nСводная таблица средней цены по бренду и полу:")
print(avg_price_pivot)
```
